```python
# Calculate factorial
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)

# Test the function
result = factorial(5)
print(f"Factorial of 5 is: {result}")
```

```python
# Fibonacci sequence
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Generate first 10 fibonacci numbers
for i in range(10):
    print(fibonacci(i))
```
