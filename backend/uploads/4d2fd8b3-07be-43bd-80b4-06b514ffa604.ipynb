# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
tuple_list = [(6, 24, 12), (7, 9, 6), (12, 18, 21)] 
  
print("The original list is : " + str(tuple_list)) 
  
K = 6
  
res = [sub for sub in tuple_list if all(ele % K == 0 for ele in sub)] 
  
print("K Multiple elements tuples : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tupple = (5, 6, 7, 4, 2) 
  
demolist = [] 
  
k = input("Enter String: ")
  
for i in tupple:
  demolist.append(i)
  demolist.append(k)
   
print("Converted Tuple with K : ", demolist) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tuple_list = [(6, 24, 12), (7, -9, 6), (-12, 18, -21)] 
  
print("The original list is : " + str(tuple_list)) 
   
res = [sub for sub in tuple_list if all(ele > 0 for ele in sub)] 
  
print("K Multiple elements tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
Tuple= (5,9,6,2,3,4,5,6,5)
counter = 0;
checker = int(input("Enter element to check"))

for i in Tuple:
  if i == checker:
    counter+=1

print("Frequency of "+str(checker)+" = "+str(counter))
```

# 05) WAP to Convert Binary tuple to integer

```python
binaryTuple = (1,0,0,0)
intVal=0
n=len(binaryTuple)-1
for i in range(0,len(binaryTuple)):
    intVal = intVal + (pow(2,n-i)*binaryTuple[i])
    
print(intVal)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
Tuple= (5,9,6,2,3,4,5,6,5)
checker = int(input("Enter element to check"))

for i in range(0,len(Tuple)):
  if Tuple[i] == checker:
    print(str(checker)+" Element Exist at position "+str(i+1))
    break

else:
  print(str(checker)+" Element does not Exist")
```
