# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
a=(1,2,3,4,5,6,7,8)
k=int(input("Enter k:"))
for i in a:
    if i%k==0:
        print(i)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
a=(1,2,3,4,5)
b='python'
res = [j for i in a for j in (i, b)] 
print(res)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
a=[(1,2,3,4,5),(-1,2,3,-5),(10,11,12,13,14)]
res = [i for i in a if all(j >= 0 for j in i)] 
print(str(res))
```

# 04) WAP to find elements frequency in given tuple.

```python
a=[('python'),('programing'),('language'),('python'),('a')]
print(a.count('python'))
```

# 05) WAP to Convert Binary tuple to integer

```python
a=(1,1,0,1)
dec=0
for i in a:
    dec=(dec<<1)+i
print(dec)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
a=[('python','is'),('programing','language'),('used','for','machine','learning')]
if (any(input("Enter string:") in i for i in a)) : 
    print("present") 
else : 
    print("not present") 
```
