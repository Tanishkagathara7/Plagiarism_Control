# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
k = 5;
x =  [(1,2),(9,12),(65,45),(13,94),(107,222),(15,60),(110,225),(1,2),(11,25),(34,52)]
div_tuples = []
for i in x:
    if i[0]%k==0  and i[1]%k ==0:
        div_tuples.append(i)
print(div_tuples)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
y = "HelloWorld!"
x = (1,3,1,34,1,78,2,45,1,3,8,8,5,3,1,9,8,45,22)
new_list = []
for i in x:
    new_list.append(i)
    new_list.append(y)
print(new_list)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
x =  [(1,2),(9,-12),(65,45),(13,-94),(-107,222),(15,-60),(110,225),(-1,-2),(11,25),(-34,-2)]
flag = 0
pos_elements = []
for i in x:
    flag = 0
    for j in i:
        if j < 0:
            flag = 1
            break
    if flag == 0:
        pos_elements.append(i)
print(pos_elements)
```

# 04) WAP to find elements frequency in given tuple.

```python
x = (1,3,1,34,1,78,2,45,1,3,8,8,5,3,1,9,8,45,22)
distinct_elements = { i for i  in x}
for i in distinct_elements:
    print("{} : {}".format(i,x.count(i)))
```

# 05) WAP to Convert Binary tuple to integer

```python
bin_tuple = (1,1,0,1,0,1,0,0,1)
k = len(bin_tuple)-1
decimal = 0
for i in bin_tuple:
    decimal+= i*(2**k)
    k-=1
print(decimal)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
x = (1,3,1,34,1,78,2,45,1,[3,4,5],8,8,5,3,1,9,8,45,22)
flag = 0
for i in x:
    if type(i) == list:
        flag = 1
if flag == 1:
    print("List Element Present")
else:
    print("No List Element Present")
```
