# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
lot = [(6, 24, 12), (7, 9, 6), (12, 18, 21)] 

k = int(input("Enter K: "))

for i in range(0,3):
    count = 0
    for j in range(0,3):
        if (lot[i][j] % k == 0):
            count = count + 1
    if(count == 3):
        print(lot[i])
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
t = (6, 24, 12)

k = str(input("Enter K: "))

print([ele for sub in t for ele in (sub, k)])
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
lot = [(6, 24, 12), (7,-9, 6), (12, 18, 21)] 

for i in range(0,3):
    count = 0
    for j in range(0,3):
        if (lot[i][j] >= 0):
            count = count + 1
    if(count == 3):
        print(lot[i])
```

# 04) WAP to find elements frequency in given tuple.

```python
t = (1,2,3,3,3,4,4,5,6,7,7,7,7,8,9)

ele = int(input("Enter element: "))

count = 0

for a in t:
    if(ele == a):
        count = count + 1
print(count)
```

# 05) WAP to Convert Binary tuple to integer

```python
t = (1,1,0,1,0,0,1)

res = 0
for ele in t:  
    res = (res << 1) | ele
    
print(res)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
lst = [1,2,3,4,5,6]

t = (1,7,8,9,6)

for a in lst:
    for b in t:
        if(a==b):
            print(str(a) + " present")
```
