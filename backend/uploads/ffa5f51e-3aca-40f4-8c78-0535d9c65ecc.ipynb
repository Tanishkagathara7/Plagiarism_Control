# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
Number = [(6, 24, 12), (7, 9, 6), (12, 18, 21)]
print("The Number list is : " + str(Number)) 
K = 6
res = [sub for sub in Number if all(ele % K == 0 for ele in sub)] 
print("K Number : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
Number = (2, 3, 9, 4, 7) 
print("The tuple is : ", Number) 
K = "Python"
res = [ele for sub in Number for ele in (sub, K)] 
print("elements : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
Number = [(-5, 28, 1), (8, 7, 5), (-3, -8, 6), (5, 8, 4)] 
print("The  list is : " + str(Number)) 
res = [sub for sub in Number if all(ele >= 0 for ele in sub)]  
print("Positive Number : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
Number = [('Python', 1), ('Python1', 2), ('Python2', 3)] 
print ("The list is : " + str(Number)) 
res = list(map(lambda i : i[0], Number)).count('Python') 
print ("frequency element  : " + str(res)) 
```

# 05) WAP to Convert Binary tuple to integer

```python
Number = (0, 1, 0, 1, 0, 0, 1) 
print("The List is : " + str(Number)) 
res = int("".join(str(ele) for ele in Number), 2) 
print("Decimal number is : " + str(res)) 
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
Number = (("Python1", "py"), ("Python", "py")) 
print ("The list is " + str(Number)) 
if (any('Python' in i for i in Number)) : 
    print("Python is present") 
else : 
    print("Python is not present") 
```
