# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
x = [(4, 24, 12), (60, 12, 6), (12, 18, 21)]
l = []
k = int(input("Enter value of k : "))
flag = 0
for i in x :
    for j in i :
        if j % k != 0:
            flag = 1
    if flag == 0:
        l.append(i)
    flag = 0    
print(l)        
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
x = (5, 6, 7, 4, 9) 
l = []
k = input("enter string : ")
for i in x:
    l.append(i)
    l.append(k)
print(l)    
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
x = [(6, 24, 12), (60, -12, 6), (12, 18, 21)]
l = []
flag = 0
for i in x :
    for j in i :
        if j < 0:
            flag = 1
    if flag == 0:
        l.append(i)
    flag = 0    
print(l)
```

# 04) WAP to find elements frequency in given tuple.

```python
x = (5, 5, 7, 4 ,5, 9, 2, 5,2,3,7)
l = []
for i in x:
    if i not in l:
        print("count of {} is {}".format(i,x.count(i)))
        l.append(i)
```
