# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
test_list = [(5,10,15), (7, 9, 6), (12, 18, 21)] 
  
print("The original list is : " + str(test_list)) 
  
K = 5
  
res = list(filter(lambda sub: all(ele % K == 0 for ele in sub), test_list)) 
   
print("K Multiple elements tuples : " + str(res))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
test_tup = (1,2,3,4,5) 
 
print("The original tuple is : ", test_tup) 
   
K = "Ritu"
   
res = [ele for sub in test_tup for ele in (sub, K)] 
   
print("Converted Tuple with K : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
test_list = [(1,2,3), (-1,-2,-3), (4,5,-6), (4, 5,6)] 
   
print("The original list is : " + str(test_list)) 
   
res = [sub for sub in test_list if all(ele >= 0 for ele in sub)] 
   
print("Positive elements Tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter 
   
test_list = [('Word', 1), ('for', 2), ('Word', 3)] 
   
print ("The original list is : " + str(test_list)) 
 
res = Counter(i[0] for i in test_list) 
   
print ("The frequency of element is : " + str(res['Word']))
```

# 05) WAP to Convert Binary tuple to integer

```python
test_tuple = (1,2,3) 
  
print("The original tuple : " + str(test_tuple)) 
   
res = int(''.join(map(str, test_tuple))) 
  
print("Tuple to integer conversion : " + str(res))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
test_tuple = (("StructuredQueryLanguage", "Ritu"), ("CS_Portal", "best")) 
   
print ("The original tuple is " + str(test_tuple)) 
 
if (any('StructuredQueryLanguage' in i for i in test_tuple)) : 
    print("Structured Query Language is present") 
else : 
    print("Structured Query Language is not present")
```
