# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
num = [(1,12,34),(12,30,48),(15,45,90)]
K = int(input("Enter the K : "))
ans = [temp for temp in num if all(temp1 % K == 0 for temp1 in temp)]
print(str(ans))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tupleNew = (1,3,5,7,9)
user_input = input("Enter the string : ")
ans = [i for i in tupleNew for i in (i,user_input)]
print(ans)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tupleNew = [(-1,-3,-4,5),(21,13,41,5),(12,34,56,-1)]
ans = [i for i in tupleNew if all(element >=0 for element in i)]
print(ans)
```

# 04) WAP to find elements frequency in given tuple.

```python
tupleNew = [('Hello',1),('How',2),('are',3),('Hello',4)]
ans = list(map(lambda i:i[0],tupleNew)).count('Hello')
print("The Frequency of given word is ",ans)
```

# 05) WAP to Convert Binary tuple to integer

```python
tupleNew = (1,0,0,0)
ans = int("".join(str(element) for element in tupleNew),2)
print("The decimal of given binary is ",str(ans))
```
