# 04) 

```python
table = {}
tuples = tuple(input("Enter a list of tuples:- ").split())
for i in tuples:
    if i in table:
        table[i]+=1
    else:
        table[i]=1
# table.items()
for key in table:
    print(key + " appears " + str(table[key]) + " times")
```
