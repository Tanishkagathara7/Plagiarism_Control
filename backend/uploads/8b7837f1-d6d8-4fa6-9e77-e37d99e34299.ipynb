# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
input_list = [(6, 24, 12), (7, 9, 6), (12, 18, 21)]
print(input_list)
k=int(input("Enter the value of k:-"))
temp=[i for i in input_list if all( j%k==0 for j in i)]
temp
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
input_tuple=(i for i in input("Enter the value of tuple:-").split(","))
input_text=input("Enter the text you want to append betwwen elements:-")
my_list=[]
for i in input_tuple:
    my_list.append(i)
    my_list.append(input_text)
my_list
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
input_list=[]
while True:
    temp=input("Enter the tuple or if wanna finish enter end:-")
    if temp=="end":
        break
    else:
        temp=tuple(int(i) for i in temp.split(","))
        input_list.append(temp)
my_list=[i for i in input_list if all(j>0 for j in i)]
my_list
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter
my_tuple=tuple(i for i in input("Enter the value:-").split(","))
print(Counter(my_tuple))
```

# 05) WAP to Convert Binary tuple to integer

```python
my_tuple=tuple(int(i) for i in input("Enter the value in binary:-").split(","))
print(int("".join(str(i) for i in my_tuple),2))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
my_tuple=tuple(i for i in input("Enter the value of tuple:-").split(","))
check_value=input("Enter the value you want to check:-")
if check_value in my_tuple:
    print("Value {} was found in the tuple".format(check_value))
else:
    print("Value {} was not found in the tuple".format(check_value))
```
