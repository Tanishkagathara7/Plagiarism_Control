# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
list = [(5,15,35), (7, 9, 6), (12, 30, 21)] 
print("The original list is : " + str(list)) 
K = 5
res = [sub for sub in list if all(ele % K == 0 for ele in sub)] 
print("K Multiple elements tuples : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuple = (4,2,7,8,40,51) 
print("The original tuple is : ", tuple) 
K = "Sun"
res = [ele for sub in tuple for ele in (sub, K)] 
print("Converted Tuple with K : ", res)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
list = [(8,-3,-11), (3,4,7), (-8,5,0), (10,15,4)] 
print("The original list is : " + str(list)) 
res = [sub for sub in list if all(ele >= 0 for ele in sub)] 
print("Positive elements Tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter
list = [('1', '5'), ('10', '9'), ('10', '4')]
print("Given list of tuples : " ,list)
Frequency = Counter(i[0] for i in list)['10']
print("The frequency of element is : ",Frequency)
```

# 05) WAP to Convert Binary tuple to integer

```python
tuple = (1,1,0,0,0,0,1)
print("The original tuple is : " + str(tuple)) 
res = int("".join(str(ele) for ele in tuple), 2)  
print("Decimal number is : " + str(res))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tuple = (1,5,7,4,0,3)
print("The original tuple is : " + str(tuple))
list = [4,7,8,9,2]
res = False
for i in list:
  if i in tuple :
    res = True
    break 
print("Is any list element present in tuple : " + str(res))
```
