# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
tuple_list = [(5, 15, 25), (12, 6, 24), (2, 6, 48)] 

print("A list of tuples are : " + str(tuple_list)) 

K = int(input("Enter Number = "))

res = [sub for sub in tuple_list if all(ele % K == 0 for ele in sub)] 

print(K," Multiple elements tuples : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuple_list = tuple(input("Enter list of tuples space-separeted : ").split())
 
print("A list of tuples are : " + str(tuple_list)) 

K = input("Enter any string = ")

res = [ele for sub in tuple_list for ele in (sub, K)] 

print("Converted Tuple with string : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tuple_list = [(0, -94, 50), (1, 6, 33), (46, 88), (10, 20, 30), (-5, 66)] 

print("A list of tuples are : " + str(tuple_list)) 

res = [sub for sub in tuple_list if all(ele >= 0 for ele in sub)] 

print("Positive elements tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
frequency = {}

tuple_list = tuple(input("Enter list of tuples space-separeted : ").split())

for i in tuple_list:
    if i in frequency:
        frequency[i]+=1
    else:
        frequency[i]=1

for key in frequency:
    print(key + " appears " + str(frequency[key]) + " times")
```

# 05) WAP to Convert Binary tuple to integer

```python
tuple_list = tuple(input("Enter a list of binary tuples space-separeted : ").split())

print("A list of tuples are : " + str(tuple_list)) 

res = int("".join(str(ele) for ele in tuple_list), 2) 

print("Conversion of binary tuple to integer is : " + str(res))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tuple_list = (0, 1, 4, 6, 10, 15, 19, 23, 28, 31)

print("A list of tuples are : " + str(tuple_list)) 

list_element = [4, 17, 14, 28]

print("A list of elements are : " + str(list_element))

for i in list_element:
    check = False
    for j in tuple_list:
        if i==j:
            print(i, "is present in tuple")
            check = True
            break
    if check==False: 
        print(i, "is not present in tuple")
```
