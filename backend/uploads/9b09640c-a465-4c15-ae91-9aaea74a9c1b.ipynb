# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
data = (5,6,7,2,4,56,23,67,32,98,66,33,15,2,4)
K = int(input('Enter the Value of K : '))
for i in data:
    if int(i)%K == 0:
        print('{}\t'.format(i), end='')
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
data = ("red", "purple", "green", "white")
output = []
stri = input('Enter the string you want to add : ')
for i in data:
    output.append(i)
    output.append(stri)
print(output)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
data = (-3,-4,2,6,3,-2,-9,-5,2,-4,2,-6)

for i in data:
    if int(i) >= 0:
        print('{}\t'.format(i), end='')
```

# 04) WAP to find elements frequency in given tuple.

```python
data = (1,1,2,2,3,3,3,3,3,5,5,5,6,6,6,9)
visited = []
count = 0
for i in range(len(data)):
    if data[i] not in visited:
        for j in range(i,len(data)):
            if data[i] == data[j]:
                count += 1
        visited.append(data[i])
        print(data[i],' has frequence of : ',count)
    count = 0
```
