# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
listOfTuples = [(6, 24, 12), (12, 18, 21),(300, 15, 30), (7, 9, 6)]
K = 3
divison = [i for i in listOfTuples if all(digit % K == 0 for digit in i)]
print("Divisible by K element Tuple(s) :" + str(divison))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuple_elements = (7,10,0,15,25,35,45,65)
string = str(input("Enter String : "))
convertedTuple = [i for j in tuple_elements for i in (j,string)]
print("Converted Tuple : ",convertedTuple)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
listOfTuple = [(7,-10,0),(7,-7),(3, 6, 5),(-2,0,-7),(5, 4, 3),(7, 10)]
positive = [i for i in listOfTuple if all(elements >= 0 for elements in i)]
print("Positive Elements Tuple :",positive)
```

# 04) WAP to find elements frequency in given tuple.

```python
listOfTuple = [('ABC', 10),('DIET', 'ABC', 20),('PQR', 'STU', 30),('DIET', 40),(50, 60, 'DIET')]
frequency = list(map(lambda i : i[0] , listOfTuple)).count('DIET')
print("The Frequency Of Element Is :",frequency)
```

# 05) WAP to Convert Binary tuple to integer

```python
binaryTuple = (1,1,1,1,1,1,1,1)
integer = int("".join(str(element) for element in binaryTuple),2)
print("Integer Number is",integer)
```
