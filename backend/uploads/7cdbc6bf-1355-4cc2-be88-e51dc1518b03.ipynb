# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
tup = [(22, 8, 340), (7, 212, 30), (312, 18, 21), (26, 44, 58)] 
print(str(tup))
K = int(input('Enter value of K: '))
ans = [i for i in tup if all(j % K == 0 for j in i)]
print(ans)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
print("Please Enter comma seperated values.")
tup=tuple([(i) for i in input().split(',')]);
print(tup)
add=input("Enter string which you want to add after every elements: ")
ans=[i for j in tup for i in (j,add)]
print(ans)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tup = [(1, -71, 50), (-3, 76, 59), (52, 7, 74), (10, 3)]
print(tup)
ans = [i for i in tup if all(j>=0 for j in i)]
print(ans)
```

# 04) WAP to find elements frequency in given tuple.

```python
print("Plaese enter comma seperated values.")
tup=tuple([int(i) for i in input().split(',')]);
print(tup)
f=0
for i in tup:
    if tup.index(i)==f:
        print(i,'==>',tup.count(i))
    f+=1
```

# 05) WAP to Convert Binary tuple to integer

```python
print("Plaese enter comma seperated values.")
tup=tuple([i for i in input().split(',')])
ans = int("".join(str(i) for i in tup), 2)
print(ans)
```

# 06) WAP to check given any list elements is present in given Tuple or not.
