# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
#Our List
list = [(3,2,9),(6,4,8),(-18,-6,12)]

#Print All List
print("List : " + str(list))

#K define
K = int(input("Enter Value Of K : "))

#All Tuple Get And Check If Divisible By 6
result = [ sub for sub in list if all(element % K == 0 for element in sub) ]

#Print After Check K Divisible
print("Tuples Is Divisible By "+ str(K)+" : " + str(result))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
#Tuple 
tupl = (5, 6, 7, 4, 9) 

#Print Tuple
print("The original tuple is : ", tupl) 
  
#Enter Our Adding String After All Element
K = str(input("Enter Our String : "))
  
#All Tuple Get and After One Data Add String
res = [ele for sub in tupl for ele in (sub, K)] 
  
#Print Result
print("Converted Tuple With String Adding : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
#Our List
list = [(3,2,9),(6,4,8),(-18,-6,12)]

#Print All List
print("List : " + str(list))

#All Tuple Get And Check If Divisible By 6
result = [ sub for sub in list if all(element >= 0 for element in sub) ]

#Print After Check K Divisible
print("Positve Tuples Is "+":"+ str(result))
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter 
  
#Tuple List
tupl_list = [('Hello', 1), ('Data', 2), ('Hello', 3)] 
  
#Just Print Original Tuples List
print ("List Is : " + str(tupl_list)) 
  
# Use Counter() and List For Compare Data
res = Counter(i[0] for i in tupl_list)
  
#Result Is
print ("The Frequency Of Element Is : " + str(res['Hello'])) 
```

# 05) WAP to Convert Binary tuple to integer

```python
# initializing tuple 
tupl = (1,1) 
  
# printing original tuple 
print("Tuple is : " + str(tupl)) 

res = 0
for ele in tupl:
    res = (res << 1) | ele

# printing result  
print("Decimal number is : " + str(res))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
#Tuple
tupl = (("Hello", "World"), ("Nice", "Python")) 
  
#Print Tuple
print ("Tuple is " + str(tupl)) 

#Get Data From User
string = str(input("Enter Your Search String : "))

#any() function for check data Is Available Or Not In Tuple
# to test for value in tuple of tuple 
if (any(string in i for i in tupl)) : 
    print(string+" is present") 
else : 
    print(string+" is not present")
```
