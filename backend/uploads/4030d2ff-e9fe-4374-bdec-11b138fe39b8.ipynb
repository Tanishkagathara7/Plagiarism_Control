# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
tup = [(12,15),(42,48),(41,16),(15,12),(22,55),(12,19)]
number = int(input('Enter number: '))
count = 0;

for a,b in tup:
    if(a%number==0 and b%number==0):
        print(str(a)+','+str(b))
        count=count+1
if count==0:
    print('No element in tuple is divisible by {}'.format(number))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tup = tuple(input('Enter tuple: '))
string = str(input('Enter string: '))
li=[]

for a in tup:
    li.append(a+string)
print(li)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tup = [(-2,22),(32,44),(85,-35),(-15,12),(45,59),(-12,-99)]
count = 0;

for a,b in tup:
    if(a>0 and b>0):
        print(str(a)+','+str(b))
        count=count+1
if count==0:
    print('No element in tuple is positive')
```

# 04) WAP to find elements frequency in given tuple.

```python
tup = tuple(input('Enter tuple: '))
number = str(input('Enter element: '))
count=0

for i in tup:
    if(i==number):
        count=count+1
        
print('{} frequency is {}'.format(number,count))
```

# 05) WAP to Convert Binary tuple to integer

```python
tup = [(1010),(100010),(101010),(10101010)]
li=[]

for i in tup:
    li.append(int(str(i),2))
print(li)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tup = tuple(input('Enter tuple: '))
li = list(input('Enter list: '))
print('\n')

for i in range(0,len(li)):
    for j in range(0,len(tup)):
        count=0
        if(tup[j]==li[i]):
            count+=1
        if count!=0:
            print('list element {} found in tuple at position {}'.format(li[i],j+1))
```
