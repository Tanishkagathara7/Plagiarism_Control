# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
listtuple=[(11,24,-12),(20,-22,38),(17,29,37),(-16,44,58),(99,75,27)]
count=True
k=int(input("Enter the number :- "))
for a,b,c in listtuple :
    if(a%k==0 and b%k==0 and c%k==0):
        print("("+str(a)+","+str(b)+","+str(c)+")")
        count=False
if(count):
    print("There is no tuple which all element divisible by "+ str(k))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tup=('hello','how','are','you')
lis=[]
string=str(input("Enter the word :- "))
for i in tup:
    lis.append(i+string)
print(lis)    
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
listtuple=[(11,24,-12),(20,-22,38),(17,29,37),(-16,44,58),(-99,75,27),(4,5,-9),(5,6,8)]
count=True

for a,b,c in listtuple :
    if(a>0 and b>0 and c>0):
        print("("+str(a)+","+str(b)+","+str(c)+")")
        count=False
if(count):
    print("There is no tuple which all elements are positive. ")
```
