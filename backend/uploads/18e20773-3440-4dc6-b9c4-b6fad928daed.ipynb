# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
numbers =  [(6, 24, 12), (60, 12, 6), (12, 18, 21)]
num = ()
count = 0
k = 6

for i in numbers:
    for j in i:
        if j%k == 0:
            count+=1
            continue
        else:
            break
    if count == len(i):
        num+=(i,)
    count = 0
            
        
print(num)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
numbers =  [(-6, 24, -12), (6, 17, 21), (-1, 18, 13), (3, 4, 5)]
num = ()
count = 0
k = 6

for i in numbers:
    for j in i:
        if j > 0:
            count+=1
            continue
        else:
            break
    if count == len(i):
        num+=(i,)
    count = 0
            
        
print(num)
```

# 05) WAP to Convert Binary tuple to integer

```python
import math

binary = (1, 0, 0, 1)
pow = len(binary)-1
dec = 0

for i in binary:
    if i == 1:
        dec += math.pow(2, pow)
    pow-=1
    
print("Integer:",dec)
```
