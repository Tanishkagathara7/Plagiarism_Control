# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
test_list = [(6, 24, 12), (7, 9, 6), (12, 18, 21)]
temp = ()
print("The original list is : " + str(test_list))
K = int(input("Enter the number: "))
for i in test_list:
    temp = ()
    for j in i:
        if j % K == 0:
            temp = temp + (j,) 
        if temp == i:
            print("K Multiple elements tuples : " + str(i))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
Tuple = ()
Total_Elements = int(input("Enter the total number of elements of tuple: "))
for i in range(0,Total_Elements):
    Input = int(input("Enter element: "))
    Tuple = Tuple + (Input,)
print("The original tuple is : ", Tuple)
String = input("Enter the string: ")
List = []
for i in Tuple:
    List.append(i)
    List.append(String)
print("Converted Tuple with String : ", List) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
test_list = [(6, -24, 12), (7, 9, 6), (12, -18, 21)]
print("The original list is : " + str(test_list))
for i in test_list:
    temp = ()
    for j in i:
        if j > 0:
            temp = temp + (j,)
        if temp == i:
            print("K Multiple elements tuples : " + str(i))
```

# 04) WAP to find elements frequency in given tuple.

```python
Tuple = ()
Total_Elements = int(input("Enter the total number of elements of tuple: "))
for i in range(0,Total_Elements):
    Input = input("Enter element: ")
    Tuple = Tuple + (Input,)
Element = input("Enter the element for which you want to find the frequency: ")
Count = 0
for i in Tuple:
    if i == Element:
        Count = Count + 1
print(Count)        
```
