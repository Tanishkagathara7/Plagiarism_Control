# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
list = [(2, 2004, 4), (23, 3, 2001), (2, 3, 2004)]  
K = 2
data = [r for r in list if all(ele % K == 0 for ele in r)]  
print("Divisible By K Elements of Tuples : " + str(data)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
t = (23, 3, 2001, 2, 3, 2001) 
print("Original Tuple : ", t)  
string = "RK"
data = [k for r in t for k in (r, string)] 
print("Updated Tuple : ", data) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
list = [(2, 3, 2004), (23, 3, 2001), (-3, 5, 6), (-4, 6)] 
data = [r for r in list if all(element >= 0 for element in r)] 
print("Positive elements of Tuples : " + str(data)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter
list = [('23', '3'), ('2', '3'), ('23', '3'),('2','3')]
print("Tuple : " ,list)
freq = Counter(i[0] for i in list)['2']
print("Frequency of Element : ",freq)
```

# 05) WAP to Convert Binary tuple to integer

```python
tup = (1,0) 
print("Binary : " + str(tup)) 
data = int("".join(str(ele) for ele in tup), 2)   
print("Integer : " + str(data))  
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tup = (23, 3, 2001) 
print("Tuple : " + str(tup)) 
list = [2] 
data = False 
for i in list: 
    if i in tup : 
        data = True 
        break 
print("Is Any List Element Available In Tuple : " + str(data))  
```
