# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
listOfTuples = [(7,9,3),(16,11,17),(12,20,24)]
k=4

res  = [sub for sub in listOfTuples if all(ele % k ==0 for ele in sub)]

print("tuple divisible by K is : ",str(res))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
demoTuple = (1,2,3,4,5)

demoList = [] 

k = "added"

for i in demoTuple:
    demoList.append(i)
    demoList.append(k)
    
print(demoList)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
listOfTuples = [(-7,9,3),(16,11,17),(-12,20,-24)]

res  = [sub for sub in listOfTuples if all(ele>0 for ele in sub)]

print("tuple having only positive elements is : ",str(res))
```

# 04) WAP to find elements frequency in given tuple.

```python
demoTuple = (5,9,6,2,6,4,8,5,7,5)
counter = 0;
checker = int(input("Enter element to check frequency: "))

for i in demoTuple:
    if i == checker:
        counter+=1
    
print("Frequency of "+str(checker)+" = "+str(counter))
```

# 05) WAP to Convert Binary tuple to integer

```python
binaryTuple = (1,1,0,1,0,0,1)
intVal=0
n=len(binaryTuple)-1
for i in range(0,len(binaryTuple)):
    intVal = intVal + (pow(2,n-i)*binaryTuple[i])
    
print(intVal)
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
demoTuple = (1,2,3,4,5)
checker = int(input("Enter element you want to check: "))

for i in range(0,len(demoTuple)):
    if demoTuple[i] == checker:
        print(str(checker)+" exists in the list at position "+str(i+1))
        break
        
else:
    print(str(checker)+" does not exist in the tuple")
```
