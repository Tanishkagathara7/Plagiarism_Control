# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
test_list = [(6, 24, 12), (7, 9, 6), (12, 18, 21)] 
print("The original list is : " + str(test_list)) 
k = int(input("enter K : "))
l=[]
for sub in test_list:
    if all(ele % k==0 for ele in sub):
        l.append(sub)
print(str(l))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
test_tuple =(6, 24, 12) 
k = input("enter string : ")
print("The original tuple is : ",test_tuple) 
res = [ele for sub in test_tuple for ele in (sub, k)] 
print(res)

    

```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
test_list = [(6, -24, 12), (7, 9, 6), (12, 18, 21)] 
print("The original list is : " + str(test_list)) 

l=[]
for sub in test_list:
    if all(ele > 0 for ele in sub):
        l.append(sub)
print(str(l))
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter 

test_list = [('c', 1), ('b', 2), ('c', 3)] 

res = Counter(i[0] for i in test_list) 
print ("The frequency of element is : " + str(res['c'])) 
```
