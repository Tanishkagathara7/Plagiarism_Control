# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
tuples = [(2,4,6),(3,6,9),(8,10,12),(12,15,18)] 

K = 3
result = [i for i in tuples if all(j % K == 0 for j in i)] 
  

print(result)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuples = (1,2,3,4)
list = []
string = "python"

for i in tuples :
    for j in (i,string) :
        list.append(j)
        
print(list)        
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tuples = [(2,4,6),(-3,6,-9),(8,10,12),(-12,-15,18)] 

result = [i for i in tuples if all(j >=0 for j in i)] 
        
print(result)
```

# 04) WAP to find elements frequency in given tuple.

```python
tuples = (1,2,3,4,5,2,4,3,5,1,2,3,4)

for i in range(1,max(tuples)+1) :
    count = tuples.count(i)
    print(i,":",count,"times")
```

# 05) WAP to Convert Binary tuple to integer

```python
tuples = (1,1,0)

result = int("".join(str(i) for i in tuples), 2)  
    
print(result)    
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
list = [1,2,3,4,5,6]
tuple = (6,7,8,9,0)

elementToFind = int(input("Enter Element:"))

for i in list:
    if elementToFind == i :
        temp = i
        
for i in tuple:
    if elementToFind == i :
        temp1 = i
        
if temp == temp1 :
    print(elementToFind,"is present in tuple")
else :
    print(elementToFind,"is not present in tuple")
```
