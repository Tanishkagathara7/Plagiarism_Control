# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
list1 = [(6, 24, 12), (12, 18, 21)]   
K = 6
res = [sub for sub in list1 if all(ele % K == 0 for ele in sub)] 
print("K Multiple elements tuples : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
list1 = (29, 11, 17)  
K = "k29"  
res = [ele for sub in list1 for ele in (sub, K)]
print("Converted Tuple with K : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
list1 = [(4, 5, 9), (-3, 2, 3), (4, 6)]   
res = [sub for sub in list1 if all(ele >= 0 for ele in sub)]
print("Positive elements Tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter 
list1 = [('keyur', 1), ('be', 2), ('better', 3)]
res = Counter(i[0] for i in list1) 
print ("frequency of element is : " + str(res['keyur']))
```

# 05) WAP to Convert Binary tuple to integer

```python
list1 = (1, 2, 5) 
res = int(''.join(map(str, list1)))   
print("Tuple to integer conversion : " + str(res))
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
list1 = (("bebetter", "ka"), ("beunique", "best")) 
if (any('bebetter' in i for i in list1)) : 
    print("bebetter is present") 
else : 
    print("bebetter is not present")
```
