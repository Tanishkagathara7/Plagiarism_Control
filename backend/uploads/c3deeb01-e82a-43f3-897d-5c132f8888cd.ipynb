# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
test_1 = [(6, 24, 12), (7, 9, 6), (12, 18, 21)] 
print("The original list is :- " + str(test_1)) 
K = 6
res = [sub for sub in test_1 if all(ele % K == 0 for ele in sub)] 
  
print("K Multiple elements tuples :- " + str(res))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
test_1 = (1, 5, 8, 34, 9, 2, 27, 56)   
print("The original tuple is :- ", test_1) 
x = "abcd"

  
# list comprehension for nested loop for flatten 
res = [ele for sub in test_1 for ele in (sub, x)] 
  
# printing result 
print("Converted Tuple with x :- ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
test_1 = [(2, -55, 9), (8, 22, 37), (-4, 9, 16), (4, -6),(22, -23, -77),(-44, -1, 55)] 
print("The original list is : " + str(test_1)) 
res = [sub for sub in test_1 if all(ele >= 0 for ele in sub)] 
  
print("Positive elements Tuples :- " + str(res)) 
```

# 05) WAP to Convert Binary tuple to integer

```python
test_1 = (1, 1, 1, 1, 0, 0, 1,) 
print("The Binary tuple is :- " + str(test_1))  
res = int("".join(str(ele) for ele in test_1), 2)  
 
print("integer number is :- " + str(res))  
```
