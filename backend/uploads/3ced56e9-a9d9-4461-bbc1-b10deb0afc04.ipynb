# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
test_list = [(6, 24, 12), (7, 9, 6), (12, 18, 21)] 
print("The original list is : " + str(test_list)) 
K = 6
  
res = [sub for sub in test_list if all(ele % K == 0 for ele in sub)] 
print(K, end=" ")
print("Multiple elements tuples : " + str(res))  
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
test_tup = (5, 6, 7, 4, 9)  
string = input("Enter the string : ") 
# list comprehension for nested loop for flatten 
res = [ele for sub in test_tup for ele in (sub, string)] 
print("Converted Tuple with string : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
test_list = [(4, 5, 9), (-3, 2, 3), (-3, 5, 6), (4, 6)]  
print("The original list is : " + str(test_list)) 
  
# all() to check each element 
res = [sub for sub in test_list if all(ele >= 0 for ele in sub)] 
  
# printing result 
print("Positive elements Tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter 
test_tup = ('python', 1), ('for', 2), ('python', 3)  
print ("The original tuple is : " + str(test_tup))  
res = Counter(i[0] for i in test_tup) 
print ("The frequency of element is : " + str(res['python'])) 
```

# 05) WAP to Convert Binary tuple to integer

```python
test_tup = (1, 1, 0, 1, 0, 0, 1)  
print("The original tuple is : " + str(test_tup)) 
  
# using int() with base to get actual number 
res = int("".join(str(ele) for ele in test_tup), 2)  
print("Decimal number is : " + str(res)) 
```
