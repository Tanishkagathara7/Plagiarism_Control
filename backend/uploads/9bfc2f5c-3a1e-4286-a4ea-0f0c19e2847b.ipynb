# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
cases = [(5, 10, 20), (7, 49, 91), (16, 80, 88)]
print(cases)
k = int(input("Enter divisor k = "))
print([i for i in cases if all(j % k == 0 for j in i)])
```

# 02) WAP to convert tuple into list by adding the given string after every element.

```python
tuple = ("DARSHAN", "MIT", "NIT", "IIT")
print("Tuple =",tuple)
list=[]
word = "COLLEGE"
for i in tuple:
    list.append(i)
    list.append(word)
tuple = list
print("List =", tuple)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
cases = [(5, -10, 20), (7, 49, -91), (9, 19, 29), (126, 74, 30), (-16, 80, 88)]
print(cases)
print("POSITIVE TUPLES ARE AS BELOW :")
print([i for i in cases if all(j>0 for j in i)])
```

# 04) WAP to find elements frequency in given tuple.

```python
tuple = (10, 8, 5, 2, 10, 15, 10, 8, 5, 8, 8, 2)
print(tuple)
find = int(input("Enter element - "))
count = 0
for i in tuple: 
    if (i == find): 
        count = count + 1
print("Frequency of",find,"=",count,"times")
```

# 05) WAP to Convert Binary tuple to integer

```python
tuple = (1,1,0,1)
print("Bianry Tuple = ",tuple)
print("Integer value = ",int("".join([str(i) for i in tuple]),2))
```
