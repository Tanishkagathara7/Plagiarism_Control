# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
K = int(input("Enter a Value of K "))
print(tuple(i for i in tuple(input("Enter a list of tuples space-sapareted").split()) if int(i) % K == 0))
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
string  = input("Enter your string ")
print(list(i + string for i in tuple(input("Enter list of tuples space - saparated : ").split())))
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
print(tuple(i for i in tuple(input("Enter a list of tupels ").split()) if int(i) > 0))
```

# 04) WAP to find elements frequency in given tuple.

```python
table = {}
tuples = tuple(input("Enter a list of tuples:- ").split())
for i in tuples:
    if i in table:
        table[i]+=1
    else:
        table[i]=1
# table.items()
for key in table:
    print(key + " appears " + str(table[key]) + " times")
```

# 05) WAP to Convert Binary tuple to integer

```python
print(int("".join(str(i) for i in tuple(input("Enter a binary tuple space-sapareted ").split())), 2) )
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tuples = tuple(input("Enter list of elememts: ").split())
element = input("Enter a element to check ")
for i in tuples:
    if i == element:
        print("Found")
        break
else: 
    print("Not Found")
```
