# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
listOne = [(5,10,15), (3,6,9), (4,24,44)]
k=5
for tempTuples in listOne :
    count = 0
    for element in tempTuples :
        if element % k == 0 :
            count = count + 1
    if count == len(tempTuples) : 
        print(tempTuples)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tupleOne = (3, 5, 7)
tempString = 'abc'
listTwo = []
for element in tupleOne :
    listTwo.append(element)
    listTwo.append(tempString)
listTwo
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
listThree = [(1,19,22), (7,-3,74), (5,94,23)]
for tempTuples in listThree :
    count = 0
    for element in tempTuples :
        if element > 0 :
            count = count + 1
    if count == len(tempTuples) : 
        print(tempTuples)
```

# 04) WAP to find elements frequency in given tuple.

```python
tupleTwo = (5, 6, 7, 9, 9, 7, 6, 2, 7, 6, 3, 4, 6)
number = int(input('Enter number which you want to check frequency : '))
count = 0
for element in tupleTwo :
    if element == number :
        count = count + 1
print('The frequency of', number, 'is', count)
```

# 05) WAP to Convert Binary tuple to integer

```python
tupleThree = (1,0,1,0,1,0)
count = len(tupleThree)
binaryNumber = 0
for element in tupleThree :
    count = count - 1
    binaryNumber = binaryNumber + element * pow(2,count)
binaryNumber
```

# 06) WAP to check given any list elements is present in given Tuple or not.
