# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
k=int(input('Enter no to be divided'))
n=int(input('Enter no of elements in the tuple'))
tuplelist=[]
for i in range(1,n):
    tuplelist.append(i)
tupleexample=tuple(tuplelist)    
flag=0
req=""
for elements in tupleexample:
    if(elements % k==0):
        flag=1
        req += str(elements) + " "
if flag==1:
    print('Elements divisible by',k,'are :',req)
else:
    print('No element is the the list divisible by k')
```

# 04) WAP to find elements frequency in given tuple.

```python
random=(1,1,1,2,2,2,5,5,6,6,7,9,0,4)
req=[]
count=0
for elements in range(len(random)):
    if random[elements] not in req:
        for repeats in range(elements-1,len(random)):
            if random[elements]==random[repeats]:
                count+=1
        req.append(random[elements])
        print(random[elements],'has a frequency of : ',count)
    print(req)    
    count=0    
```

# 05) WAP to Convert Binary tuple to integer

```python
binary=(1,0,1,1)

ans=0
count=0
for num in binary[::-1]:    
    ans += 2**count*num
    count+=1
print('Integer to binary is:',ans)    
```
