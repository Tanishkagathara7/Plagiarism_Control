```python
# Calculate factorial - very similar to student 1
def calculate_factorial(num):
    if num <= 1:
        return 1
    else:
        return num * calculate_factorial(num-1)

# Test the function
answer = calculate_factorial(5)
print(f"Factorial of 5 is: {answer}")
```

```python
# Fibonacci sequence - similar logic
def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)

# Generate first 10 fibonacci numbers
for j in range(10):
    print(fib(j))
```
