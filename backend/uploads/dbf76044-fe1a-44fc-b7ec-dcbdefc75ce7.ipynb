# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
Numberlist = [(6, 24, 12), (7, 9, 6), (12, 18, 21)]
print("The list is : " + str(Numberlist)) 
K = 6
res = [sub for sub in Numberlist if all(ele % K == 0 for ele in sub)] 
print("Kelements : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
Numberlist = (5, 6, 7, 4, 9) 
print("The tuple is : ", Numberlist) 
K = "PythonWorld"
res = [ele for sub in Numberlist for ele in (sub, K)] 
print("elements : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
Numberlist = [(4, -8, 1), (3, 2, 3), (-3, -8, 6), (4, 7,6)] 
print("The  list is : " + str(Numberlist)) 
res = [sub for sub in Numberlist if all(ele >= 0 for ele in sub)]  
print("Positive elements : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
Numberlist = [('Python', 1), ('for', 2), ('EveryThing', 3)] 
print ("The original list is : " + str(Numberlist)) 
res = list(map(lambda i : i[0], Numberlist)).count('Python') 
print ("frequency element  : " + str(res)) 
```

# 05) WAP to Convert Binary tuple to integer

```python
Numberlist = (1, 1, 0, 1, 0, 0, 1) 
print("The List is : " + str(Numberlist)) 
res = int("".join(str(ele) for ele in Numberlist), 2) 
print("Decimal number is : " + str(res)) 
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
Numberlist = (("PythonForEverything", "pfe"), ("Python", "best")) 
print ("The list is " + str(Numberlist)) 
if (any('PythonForEverything' in i for i in Numberlist)) : 
    print("PythonForEverything is present") 
else : 
    print("PythonForEverything is not present") 
```
