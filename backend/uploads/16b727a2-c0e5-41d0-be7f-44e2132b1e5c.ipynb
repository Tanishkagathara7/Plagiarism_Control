# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
print("Enter no. of tuples you want to enter")
n=int(input())
print("Enter value of K")
k=int(input())
lis = []
lisans=[]
flag=0
for i in range(0,n):
  print("Enter tuple elements")
  lis.append(tuple(map(int,input().split())))
for i in lis:
  flag=0
  for j in i:
    if j%k==0:
      continue
    else:
      flag=1
      break
  if flag==0:
    lisans.append(i)
print(lisans)
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
print("Enter Tuple elements")
tup = tuple(input().split())
print("Enter string to add")
s=input()
ans=[]
for i in tup:
  ans.append(i)
  ans.append(s)

print(ans)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
print("Enter no. of tuples you want to enter")
n=int(input())
lis = []
lisans=[]
flag=0
for i in range(0,n):
  print("Enter tuple elements")
  lis.append(tuple(map(int,input().split())))
for i in lis:
  flag=0
  for j in i:
    if j>=0:
      continue
    else:
      flag=1
      break
  if flag==0:
    lisans.append(i)
print(lisans)
```

# 04) WAP to find elements frequency in given tuple.

```python
tup = tuple(map(int,input().split()))
lis=[0]*100
for i in tup:
  lis[i]+=1
print("Frequency is :")
for i in range(0,100):
  if lis[i]>0:
    print(i,"-> ",end="")  
    print(lis[i])
```

# 05) WAP to Convert Binary tuple to integer

```python
test_tup = tuple(map(int,input().split())) 

res = 0
for i in test_tup:  

    res = (res << 1) | i  
print("Decimal number is : " + str(res))  
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
print("Enter tuple")
tup = tuple(map(int,input().split()))
print("Enter list")
lis= list(map(int,input().split()))
f=0
for i in range(0,len(tup)-len(lis)):
  for j in range(0,len(lis)):
    if tup[i+j]!=lis[j]:
      break
  if j==len(lis)-1:
    print("List element is Present in Tuple")
    f=1
    break
if f==0:
   print("List element is Not Present in Tuple")
```
