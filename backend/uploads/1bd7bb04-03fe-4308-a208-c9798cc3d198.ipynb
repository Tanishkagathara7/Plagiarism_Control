# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
list = [(8, 20, 2), (9, 3, 12), (21, 11, 24)]  
print("The original list is : " + str(list)) 
K = 2
res = [sub for sub in list if all(ele % K == 0 for ele in sub)]  
print("Multiple elements tuples which have all elements divisible by " + str(K) + " : " + str(res)) 
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuple1 = (1, 3, 5, 7, 9) 
print("The original tuple is : ", tuple1)  
string = "Keval"
res = [ele for sub in tuple1 for ele in (sub, string)] 
print("Converted Tuple with string : ", res) 
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
tuple_list = [(1, -3, -5), (4, 8, 12), (2, -4, 6), (3, 6, 9)] 
print("The original list is : " + str(tuple_list)) 
res = [sub for sub in tuple_list if all(ele >= 0 for ele in sub)] 
print("Positive elements of Tuples : " + str(res)) 
```

# 04) WAP to find elements frequency in given tuple.

```python
from collections import Counter
tuple_list = [('2', '3'), ('4', '2'), ('4', '6'), ('7', '9')]
print("Given list of tuples : " ,tuple_list)
Frequency_res = Counter(i[0] for i in tuple_list)['4']
print("The frequency of element is : ",Frequency_res)
```

# 05) WAP to Convert Binary tuple to integer

```python
binary_tuple = (1, 0, 1, 1, 1, 0, 1) 
print("The original tuple is : " + str(binary_tuple)) 
res = int("".join(str(ele) for ele in binary_tuple), 2)   
print("integer number is : " + str(res))  
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tuple_list = (1, 3, 5, 7, 9, 11, 13) 
print("The original tuple is : " + str(tuple_list)) 
list1 = [1, 11, 7] 
res = False 
for i in list1: 
    if i in tuple_list : 
        res = True 
        break 
print("Is any list element present in tuple : " + str(res))  
```
