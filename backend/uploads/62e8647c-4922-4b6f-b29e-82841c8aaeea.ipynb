# 01) WAP to find tuples which have all elements divisible by K from a list of tuples.

```python
list1 = [(20,10,5,25,30),(5,1,7,9,10,23,4,75),(10,20,60,40,80,10030),(5,3,6,7,20,200),(9,3,6,15,21),(2,4,6,98,40)]
n= int(input("Enter Number:"))
for i in range(0,len(list1)):
    flag = 0
    for j in range(0,len(list1[i])):
        if(list1[i][j]%n!=0):
            flag=1
            break
        else:
            continue
    if(flag==0):
        print(list1[i])
```

# 02) WAP to covert tuple into list by adding the given string after every element.

```python
tuple1 = (20,5,23,6,1,42)
word = input("Enter string you want to insert after every element:")
list1=[]
counter=0
for i in range(0,len(tuple1)):
    list1.append(tuple1[counter])
    list1.append(word)
    counter=counter+1
print("Converted List:",list1)
```

# 03) WAP to find Tuples with positive elements in List of tuples.

```python
list1 = [(20,10,5,-25,30),(5,1,7,9,-10,23,4,75),(-10,20,-60,40,80,10030),(5,3,6,-7,20,200),(9,3,6,15,21),(2,4,6,98,40)]
for i in range(0,len(list1)):
    flag = 0
    for j in range(0,len(list1[i])):
        if(list1[i][j]<0):
            flag=1
            break
        else:
            continue
    if(flag==0):
        print(list1[i])
```

# 04) WAP to find elements frequency in given tuple.

```python
tuple1 = (2,8,3,6,1,10,4,50,22,7,8,1,9,10,22,5,22,10,5)
n = int(input("Enter Number:"))
if(tuple1.count(n)>0):
    print("Frequency:",tuple1.count(n))
else:
    print("Tuple Doesn't Countain Your Entered Number.")
```

# 05) WAP to Convert Binary tuple to integer

```python
tuple1 = (1, 1, 0, 1, 1)
counter=0
sum=0
for i in reversed(range(len(tuple1))):
    sum=sum+tuple1[i]*pow(2,counter)
    counter=counter+1
print("Integer:",sum)
    
```

# 06) WAP to check given any list elements is present in given Tuple or not.

```python
tuple1 = (1,8,9,20,11,52,20,1,4,6,9,7,30,5,62,34,78,90,100,12,43,53,20)
n = int(input("Enter number Of Elements Of List:"))
list1=[]
for i in range(0,n):
    a=int(input())
    list1.append(a)
print("------------------------------------------------------------")
for k in range(0,n):
    if list1[k] in tuple1:
        print(list1[k],"Is Present In Tuple.")
    else:
        print(list1[k],"Is Not Present In Tuple.")
```
